##**1. Dataset Publik**

Sumber data: https://www.kaggle.com/datasets/vipullrathod/fish-market

```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##**2. Load data ke Data Frame (df)**

```python
# menghubungkan google drive ke colab
from google.colab import drive
drive.mount('/content/drive')

# load dataset
df = pd.read_csv('/content/drive/MyDrive/Dataset PDS/Fish.csv')
df
```

##**3. Merapikan dan melihat informasi data**

```python
df.info()
```

```python
# Statistik deskriptif
df.describe()
```

##**4. Mengecek missing value**

```python
df.isnull().sum()
```

##**5. Manipulating data (membuat variabel baru)**

```python
# Membuat variabel baru
df["Body Volume"] = df["Length3"] * df["Height"] * df["Width"]
df["Weight per Length"] = df["Weight"] / df["Length3"]

df
```

##**6. Sorting**

Mengurutkan ikan berdasarkan variabel numerik utama (Weight)






```python
# Daftar 10 ikan paling berat
df.sort_values(by="Weight", ascending=False).head(10)
```

```python
# Daftar 10 ikan paling ringan
df.sort_values(by="Weight", ascending=True).head(10)
```

```python
# Mengahapus data yang tidak valid
# Spesies Roach dengan bobot 0.0 gram pada indeks ke-40
df = df.drop(40)
```

```python
# Daftar 10 ikan paling ringan (NEW)
df.sort_values(by="Weight", ascending=True).head(10)
```

##**7. Grouping**

Mengelompokkan data berdasarkan spesies dan menghitung mean, sum, serta count


```python
df.groupby("Species")["Weight"].agg(["mean","sum","count"])
```

##**8. Pivot Table**

Pivot table dibuat dengan Species sebagai index dan variabel numerik sebagai values menggunakan fungsi agregasi mean.

```python
pivot_table = df.pivot_table(
    index='Species',
    values=['Weight', 'Length1', 'Length2', 'Length3',
            'Height', 'Width', 'Body Volume', 'Weight per Length'],
    aggfunc='mean'
)

pivot_table
```

##**9. Plotting**

```python
# Mengitung rata-rata Body Volume per spesies
mean_volume = df.groupby('Species')['Body Volume'].mean()

# Plot
plt.figure()
mean_volume.plot(kind='bar')

plt.title('Rata-rata Body Volume Ikan per Spesies')
plt.xlabel('Spesies')
plt.ylabel('Rata-rata Body Volume (cm³)')

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
df.boxplot(column='Body Volume', by='Species')

plt.title('Distribusi Body Volume per Spesies')
plt.suptitle('')
plt.xlabel('Spesies')
plt.ylabel('Body Volume (cm³)')

plt.xticks(rotation=45)
plt.show()
```
